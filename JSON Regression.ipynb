```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_json('solution_table.json')
print('done')
```

```python
print (df.head(3))
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df.dtypes)
# df.to_csv('test.csv',sep='\t', encoding='utf-8')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df.head(3))
print(df.shape[0])
count_nan = df['beg_fov'].isnull().sum()
print ('Count of NaN: ' + str(count_nan))
```

```python
import numpy as np
df['elevation_norot_rad'] = np.deg2rad(df['elevation_norot'])
df['cos_elevation_norot'] = np.cos(df['elevation_norot_rad'])
df['height_ch'] = df['begin_height'] - df['end_height']
df1 = df[(df['beg_fov'] == True) & (df['end_fov'] == True) & (df['height_ch'] > 0) ]
```

```python
df2 = df1[['vel_avg_norot','cos_elevation_norot','peak_abs_mag','height_ch']].dropna()
X = df2[['vel_avg_norot','cos_elevation_norot','peak_abs_mag']]
Y = df2['height_ch']
X = sm.add_constant(X)   
model= sm.OLS(Y,X).fit()   
model.summary()
```

```python
predictions2 = model.get_prediction(X).summary_frame(alpha=0.05)
df['obs_ci_lower'] = predictions2['obs_ci_lower']
df['obs_ci_upper'] = predictions2['obs_ci_upper']
df['mean'] = predictions2['mean']
predictions2
```

```python
predictions1 = model.get_prediction(X).summary_frame(alpha=0.32)
df['obs_ci_lower1'] = predictions1['obs_ci_lower']
df['obs_ci_upper1'] = predictions1['obs_ci_upper']
predictions1
```

```python
predictions3 = model.get_prediction(X).summary_frame(alpha=0.003)
df['obs_ci_lower3'] = predictions3['obs_ci_lower']
df['obs_ci_upper3'] = predictions3['obs_ci_upper']
predictions3
```
