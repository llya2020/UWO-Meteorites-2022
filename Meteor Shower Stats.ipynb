```python
plt.figure(figsize=(30, 20)) 

for k in range (8):
    plt.subplot(3,3,k+1)
    df_bin_mag = df[(df['peak_abs_mag'] > k) & (df['peak_abs_mag'] < k+1)]
    n = df_bin_mag.shape[0]
    if (df_bin_mag.shape[0] == 0):
        continue
    loc = []
    fractions = []
    for j in range (13):
        df_section = df_bin_mag[(df_bin_mag['vel_avg_norot'] > 10 + j*5) & (df_bin_mag['vel_avg_norot'] < 15 + j*5)]
        if (df_section.shape[0] > 5):
            df_nonshower = df_section[(df_section['shower_code'] == "...")]
            frac = df_nonshower.shape[0] / df_section.shape[0]
            fractions.append(frac)
            loc.append(12.5+j*5)

    plt.bar(loc,fractions,width=5,ec="k")
    plt.xlim(0,80)
    plt.ylim(0,1.1)
    plt.ylabel('Fraction of non-shower objects',size="x-large")
    plt.xlabel('Average apparent velocity (km/s)',size="x-large")
    plt.title('Mag = ' + '(' + str(k)  +',' + str(k+1) + '), n = ' + str(n),size="x-large")

plt.savefig('nonshower_mag_velocity_0712.jpg')
```

```python
plt.figure(figsize=(20, 30)) 

for k in range (13):
    plt.subplot(5,3,k+1)
    df_bin_vel = df[(df['vel_avg_norot'] > 10 + k*5) & (df['vel_avg_norot'] < 15 + k*5)]
    n = df_bin_vel.shape[0]
    if (df_bin_vel.shape[0] == 0):
        continue
    loc = []
    fractions = []
    for j in range (8):
        df_section = df_bin_vel[(df_bin_vel['peak_abs_mag'] > j) & (df_bin_vel['peak_abs_mag'] < j+1)]
        if (df_section.shape[0] > 5):
            df_nonshower = df_section[(df_section['shower_code'] == "...")]
            frac = df_nonshower.shape[0] / df_section.shape[0]
            fractions.append(frac)
            loc.append(0.5+j)

    plt.bar(loc,fractions,width=1,ec="k")
    plt.xlim(0,8)
    plt.ylim(0,1.1)
    plt.ylabel('Fraction of non-shower objects',size="x-large")
    plt.xlabel('Peak absolute magnitude',size="x-large")
    plt.title('Velocity = (' + str(10+k*5) + ' km/s, ' + str(15+k*5)+' km/s), n = ' + str(n))

plt.savefig('nonshower_velocity_mag_0712.jpg')
```
