```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import json
with open('solution_table.json', 'r') as f:
  data = json.load(f)
df = pd.json_normalize(data)
print('done')
```

```python
print (df.head(3))
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df.dtypes)
# df.to_csv('test.csv',sep='\t', encoding='utf-8')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df.head(3))
print(df.shape[0])
count_nan = df['beg_fov'].isnull().sum()
print ('Count of NaN: ' + str(count_nan))
dftest = df[df['num_stations'] == 2]
print(dftest.shape[0])
```

```python
df1 = df[(df['beg_fov'] == True) & (df['end_fov'] == True)]
df_jacchia = df1[(df1['accel_model'] == "jacchia")]
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_jacchia.head(10))
a1 = []
for i in range(df_jacchia.shape[0]):
    a1.append(df_jacchia['accel_coeff'].iloc[i][0])
print('done')
```

```python
df_jacchia.loc[:, 'a1'] = df_jacchia.accel_coeff.map(lambda x: x[0])
df_jacchia.loc[:, 'a2'] = df_jacchia.accel_coeff.map(lambda x: x[1])
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_jacchia.head(10))
```

```python
df['a1'] = df_jacchia['a1']
df['a2'] = df_jacchia['a2']
df_jacchia['a1*a2'] = df_jacchia['a1'] * df_jacchia['a2']
df['a1*a2'] = df_jacchia['a1*a2']
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df.head(10))
```

```python
df['vel_avg_norot_uncty'] = 100*df['vel_avg_norot_err'] / df['vel_avg_norot']
df['dec_geo_uncty'] = 100*df['dec_geo_err'] / df['dec_geo']
plt.hist(df['dec_geo_err'],range=[0,1])
plt.xscale("log")
```

```python
plt.figure(figsize=(20, 30))


for k in range (7):
    plt.subplot(5,3,k+1)
    df_bin_vel = df[(df['vel_avg_norot'] > 10 + k*10) & (df['vel_avg_norot'] < 20 + k*10) & (df['vel_avg_norot_uncty'] > 10**-3) & (df['vel_avg_norot_uncty'] < 10) & (df['dec_geo_uncty'] > 10**-3) & (df['dec_geo_uncty'] < 10)]
    n = df_bin_vel.shape[0]
    if (df_bin_vel.shape[0] == 0):
        continue
    plt.hexbin(df_bin_vel['dec_geo_uncty'],df_bin_vel['vel_avg_norot_uncty'],gridsize=(100,50),xscale="log",yscale="log",cmap="turbo")
    plt.xscale("log")
    plt.yscale("log")
    plt.xlim(10**-3,10**1)
    plt.ylim(10**-3,10**1)
    plt.ylabel('Velocity uncertainty (%)',size="x-large")
    plt.xlabel('Declination uncertainty (%)',size="x-large")
    plt.title('Velocity = (' + str(10+k*10) + ' km/s, ' + str(20+k*10)+' km/s), n = ' + str(n))

#plt.savefig('velocity_declination_uncertainty_0714.jpg')
```
