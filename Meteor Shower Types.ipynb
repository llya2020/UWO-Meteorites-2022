```python
import math
j=0
plt.figure(figsize=(16,8))
df_show = df[(df['shower_code'] == countsf.index[j])]
n,bins,patches=plt.hist(df_show['peak_abs_mag'],bins=range(math.floor(df_show['peak_abs_mag'].min()), math.ceil(df_show['peak_abs_mag'].max()), 1))
plt.clf()
n = n.tolist()
bins = bins.tolist()
bins.pop()
for i in range (len(bins)):
    bins[i] = bins[i] + 0.5
n = np.cumsum(n)
plt.bar(bins,n,width=1,ec="k")
m,b = np.polyfit(bins,n,1)
plt.axline(xy1=(0, b), slope=m, label=f'$y = {m}x {b:+}$',c="red")
plt.legend(loc='best',markerscale=5)
plt.xlabel("Peak absolute magnitude",size="x-large")
plt.ylabel("Count (cumulative)",size="x-large")
plt.ylim(bottom=0)
plt.title(countsf.index[j],size="x-large")
date="_0810.jpg"
#plt.savefig("all_showers/" + countsf.index[j] + "/mag_hist" + date)
#plt.clf()
```

```python
j=0
plt.figure(figsize=(16,8))
df_show = df[(df['shower_code'] == countsf.index[j])]
count = []
n=0
mags_sorted = df_show['peak_abs_mag'].sort_values().dropna()
for k in range(len(mags_sorted)):
    n = n + 1
    count.append(n)
plt.scatter(mags_sorted,count,s=5,c="red")
plt.yscale("log")
count_nan = mags_sorted.isnull().sum()
m,b = np.polyfit(mags_sorted,np.log10(count),1)
plt.plot(mags_sorted, 10**((m*mags_sorted)+b),label=f'$\log_{{{10}}} \ y = {m}x {b:+}$')
plt.legend(loc='best',markerscale=5)
plt.ylabel("Count (cumulative)",size="x-large")
plt.xlabel("Peak absolute magnitude",size="x-large")
plt.title(countsf.index[j] + ', n = ' + str(countsf['shower_code'].iloc[j]),size="x-large")
date="_0816.jpg"
plt.xticks(range(math.floor(mags_sorted.min()), math.ceil(mags_sorted.max()), 1))
plt.grid()
#os.remove("all_showers/" + countsf.index[j] + "/mag_cumulative" + "_0814.jpg")
#plt.savefig("all_showers/" + countsf.index[j] + "/mag_cumulative" + date)
#plt.clf()
```

```python
j=0
plt.figure(figsize=(16,8))
df_show = df[(df['shower_code'] == countsf.index[j])]
count = []
n=0
mass_sorted = df_show['mass'].sort_values().dropna()
mass_sorted = mass_sorted.iloc[::-1]
for k in range(len(mass_sorted)):
    n = n + 1
    count.append(n)
plt.scatter(mass_sorted,count,s=5,c="red")
plt.yscale("log")
plt.xscale("log")
m,b = np.polyfit(np.log10(mass_sorted),np.log10(count),1)
plt.plot(mass_sorted, 10**((m*np.log10(mass_sorted))+b),label=f'$\log_{{{10}}} \ y = {m}\ \log_{{{10}}} \ x {b:+}$')
plt.legend(loc='best',markerscale=5)
plt.ylabel("Count (cumulative)",size="x-large")
plt.xlabel("Mass (kg)",size="x-large")
plt.title(countsf.index[j] + ', n = ' + str(countsf['shower_code'].iloc[j]),size="x-large")
date="_0816.jpg"
plt.grid()
#os.remove("all_showers/" + countsf.index[j] + "/mass_cumulative" + "_0814.jpg")
#plt.savefig("all_showers/" + countsf.index[j] + "/mass_cumulative" + date)
#plt.clf()
```
