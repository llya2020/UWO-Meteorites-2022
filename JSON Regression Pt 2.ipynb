```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np

df = pd.read_json('solution_table.json')
print('done')
```

```python
print (df.head(3))
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df.dtypes)
# df.to_csv('test.csv',sep='\t', encoding='utf-8')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df.head(3))
print(df.shape[0])
count_nan = df['beg_fov'].isnull().sum()
print ('Count of NaN: ' + str(count_nan))
```

```python
df['elevation_norot_rad'] = np.deg2rad(df['elevation_norot'])
df['cos_elevation_norot'] = np.cos(df['elevation_norot_rad'])
df['height_ch'] = df['begin_height'] - df['end_height']
df1 = df[(df['beg_fov'] == True) & (df['end_fov'] == True) & (df['height_ch'] > 0) & (df['F'] >= 0.666) & (df['F'] <= 1.0)]

df2 = df1[['vel_avg_norot','cos_elevation_norot','peak_abs_mag','height_ch']].dropna()
X = df2[['vel_avg_norot','cos_elevation_norot','peak_abs_mag']]
Y = df2['height_ch']
X = sm.add_constant(X)   
model= sm.OLS(Y,X).fit()   
model.summary()
```

```python
predictions = model.get_prediction(X).summary_frame(alpha=0.05)
df['mean'] = predictions['mean']
predictions
```

```python
df2['height_ch_theoretical'] = 25.5485 + 0.0549*df2['vel_avg_norot'] - 11.0002*df2['cos_elevation_norot'] - 2.1981*df2['peak_abs_mag']
df['height_ch_theoretical'] = df2['height_ch_theoretical']
df['abs_height_diff'] = (df['height_ch'] - df['height_ch_theoretical'])
df['percent_error'] = df['abs_height_diff'] / df['height_ch'] *100
```

```python
df1 = df[(df['beg_fov'] == True) & (df['end_fov'] == True) & (df['height_ch'] > 0) & (df['F'] >= 0.666) & (df['F'] <= 1.0)]
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df1.head(3))
```
